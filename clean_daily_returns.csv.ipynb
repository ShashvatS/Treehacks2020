```python
import pandas as pd
from tqdm import tqdm
import datetime
```

```python
df = pd.read_csv("fundamentals-1_cleaned.csv")
df = df.rename(columns={"tic": "TICKER", "datadate": "date"})
df = df.set_index(['TICKER', "date"])
df = df.astype('float64')
df = df.dropna(axis = "columns", thresh = 0.2 * len(df.index))
df = df.dropna(thresh = 3)
df = df.fillna(0)
df.sort_index(inplace = True)
df2 = df
```

```python
df = pd.read_csv("daily_returns.csv")
df = df.set_index(['TICKER', "date"])

for col in df.columns:
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace("C", "0.0")

ret_copy = df["RET"]
df = df.drop(columns = ["RET"])

df["newquarter"] = pd.Series(0.0, index=df.index)
for col in cols:
    df[col] = pd.Series(0.0, index=df.index)
    
df["RET0"] = ret_copy

for i in range(1, 61):
    col = "RET" + str(i)
    df[col] = pd.Series(0.0, index=df.index)
    
df = df.fillna(0)
df.sort_index(inplace = True)
```

```python
first_occur = {}

for row in tqdm(df2.iterrows()):
    idx = row[0]

    if idx[0] not in first_occur.keys():
        first_occur[idx[0]] = idx[1]
    elif idx[1] < first_occur[idx[0]]:
        first_occur[idx[0]] = idx[1]
```

```python
for row in tqdm(df2.iterrows()):
    if row[0] in df.index:
        for col in df2.columns:
            df.at[row[0], col] = df2.at[row[0], col]
        
        df.at[row[0], "newquarter"] = 1
        
df.sort_index()
df.to_csv("daily_returns2.csv")
```

```python
with open("daily_returns2.csv", "r") as f:
    first_line = True    
    out = open("daily_returns3.csv", "w")
    
    ctr = 0
    dates = {}
    ret_idx = None
    gvkey_idx = None
    nq_idx = None
    
    prev_values = None
    prev_ticker = None
    
    for line in f:
        if line == "\n":
            break
        if first_line:
            line = line.split(",")
            ret_idx = line.index("RET0")
            gvkey_idx = line.index("gvkey")
            nq_idx = line.index("newquarter")
            print(ret_idx, gvkey_idx)
            out.write(",".join(line))
            first_line = False
            continue
          
        line = line.split(',')
        if line[0] == "":
            continue
        date = line[1]
        cpy = list(line)
        
        tmp = first_occur.get(line[0], None)
        skip = False
        if not tmp or int(line[1]) < tmp:
            skip = True
        
        if date not in dates.keys():
            ctr += 1
            dates[date] = ctr
            
        ticker = line[0]
        
        if ticker != prev_ticker:
            prev_ticker = ticker
            prev_values = [str(0.0)] * len(line)
            
        
        if float(line[nq_idx]) == 0:
            for i in range(gvkey_idx, ret_idx):
                line[i] = prev_values[i]
            
        for i in range(60):
            line[ret_idx + i + 1] = prev_values[ret_idx + i]
        line[-1] += '\n'
            
        prev_values = line 
    
        line[1] = str(dates[date])
        
        if skip:
            continue
    
        out.write(",".join(line))
        
    out.close()
```
